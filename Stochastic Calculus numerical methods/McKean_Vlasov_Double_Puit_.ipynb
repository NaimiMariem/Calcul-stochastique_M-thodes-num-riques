#Simuation des solutions d'EDS:


**Equation de Langevin MacKean Vlasov Ã  double puit**


```python
import numpy as np
import matplotlib.pyplot as plt


a = -1
b = 5
m = 5
N = 10
sigma=0.1
T=10
t=np.linspace(0,T,500)
x0=np.zeros((N,len(t)))
x0[:,0]=[0 for i in range(N)]
#print(x0)

beta=0
alpha=1

def simulate_langevin_euler(t,m,b,a,N,x0,beta,alpha,sigma):
  def v_deriv(x,m,a,b):

    return 2*m*(x-a)*(x-b)*(2*x-b-a)/(((b-a)/2)**2)


  for j in range(1,len(t)):

    dt=t[j]-t[j-1]
    for i in range(1,N):

        x0[i,j]=x0[i,j-1]+beta*v_deriv(x0[i,j-1],m,a,b)*dt+alpha*(np.mean(x0[:,j-1])-x0[i,j-1])*dt+ sigma*np.sqrt(dt)*np.random.normal(0,1)
  return x0




X_L=simulate_langevin_euler(t,m,b,a,N,x0,beta,alpha,sigma)





plt.plot(t,X_L[5,:])




```

  Commentaire:

  L'allure des trajectoires changent Ã©normÃ©ment avec la valeur de $X_0$.
  Pour $\sigma$=0, et $\beta$>0, les trajectoires sont convergentes et Ã©gales sur une grande plage de valeurs.

  Le cas $\beta< 0$ est un cas rÃ©puslif, toutes les trajectoires divergent.
